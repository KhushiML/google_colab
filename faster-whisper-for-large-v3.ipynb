```python
from faster_whisper import WhisperModel

model_size = "large-v3"

# Run on GPU with FP16
# model = WhisperModel(model_size, device="cuda", compute_type="float16")

# or run on GPU with INT8
model = WhisperModel(model_size, device="cuda", compute_type="default")
# or run on CPU with INT8
# model = WhisperModel(model_size, device="cpu", compute_type="int8")
```

```python
segments, info = model.transcribe("zh.wav", beam_size=5)

print("Detected language '%s' with probability %f" % (info.language, info.language_probability))

for segment in segments:
    print("[%.2fs -> %.2fs] %s" % (segment.start, segment.end, segment.text))
```

```python
import time
start_time = time.time()

# Transcribe the audio file
segments, info = model.transcribe("/kaggle/input/hindi-data/1010977_2024.08.27_13.15.20-2024.08.27_13.16.20.mp3", beam_size=5)

# End timing
end_time = time.time()

# Calculate the processing time
processing_time = end_time - start_time

# Print detected language and its probability
print("Detected language '%s' with probability %f" % (info.language, info.language_probability))

# Print segments
for segment in segments:
    print("[%.2fs -> %.2fs] %s" % (segment.start, segment.end, segment.text))

# Print processing time
print("Processing time: %.2f seconds" % processing_time)
```
