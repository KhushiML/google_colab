```python
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load the T5 model and tokenizer
model_name = 't5-small'  # You can also use 't5-base' or 't5-large'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

def generate_headline(substory):
    # Prepend the task-specific prefix
    input_text = f"summarize: {substory}"
    
    # Tokenize the input text
    input_ids = tokenizer.encode(input_text, return_tensors='pt')
    
    # Generate the output
    with torch.no_grad():
        output_ids = model.generate(input_ids, max_length=20, num_beams=4, early_stopping=True)
    
    # Decode the generated output to get the headline
    headline = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return headline

# Example substories
substories = [
    "The government announced a new education reform aimed at improving STEM education in schools. This initiative will introduce coding classes from the elementary level.",
    "In a recent press conference, the Minister of Health revealed a new vaccination campaign targeting young children. This effort aims to increase immunization rates across the country.",
    "A major breakthrough in renewable energy technology was announced by researchers at a national lab. The new solar panel design is said to be 50% more efficient than current models.",
    "The city council is launching a new program to enhance digital literacy among students. Starting next semester, schools will offer free workshops on coding and robotics.",
    "Health officials are implementing a new strategy to ensure higher vaccination coverage among toddlers. This includes community outreach and educational seminars for parents.",
    "Scientists have developed a revolutionary new type of wind turbine that promises to generate more energy with less environmental impact. This innovation could transform the renewable energy sector."
]

# Generate standardized headlines for each substory
for substory in substories:
    headline = generate_headline(substory)
    print(f"Substory: {substory}\nGenerated Headline: {headline}\n")
```

```python
import numpy as np
from sentence_transformers import SentenceTransformer, util

# Load the SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Example substories and their standardized headlines
substories = [
    "The government announced a new education reform aimed at improving STEM education in schools.",
    "In a recent press conference, the Minister of Health revealed a new vaccination campaign targeting young children.",
    "A major breakthrough in renewable energy technology was announced by researchers at a national lab.",
    "The city council is launching a new program to enhance digital literacy among students.",
    "Health officials are implementing a new strategy to ensure higher vaccination coverage among toddlers.",
    "Scientists have developed a revolutionary new type of wind turbine."
]

# Corresponding standardized headlines
standardized_headlines = [
    "Government Unveils New Education Reform",
    "New Vaccination Campaign for Children Launched",
    "Breakthrough in Renewable Energy Technology Announced",
    "City Council Enhances Digital Literacy Program",
    "New Strategy to Boost Vaccination Rates",
    "Revolutionary Wind Turbine Developed"
]

# Encode substories and headlines
substory_embeddings = model.encode(substories, convert_to_tensor=True)
headline_embeddings = model.encode(standardized_headlines, convert_to_tensor=True)

# Function to generate standardized headlines
def generate_standardized_headline(substory):
    # Encode the new substory
    substory_embedding = model.encode(substory, convert_to_tensor=True)

    # Compute cosine similarities
    similarities = util.pytorch_cos_sim(substory_embedding, headline_embeddings)

    # Move similarities to CPU and convert to numpy
    similarities_np = similarities.cpu().detach().numpy()

    # Find the most similar headline
    most_similar_idx = np.argmax(similarities_np)
    return standardized_headlines[most_similar_idx]

# Example usage
for substory in substories:
    generated_headline = generate_standardized_headline(substory)
    print(f"Substory: {substory}\nGenerated Headline: {generated_headline}\n")
```

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer, util

# Load the SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Load your DataFrame
data = pd.read_excel('/kaggle/input/substory-headline-data/Substory_Headline_Data.xlsx')  # Ensure your CSV has 'substory' and 'headline' columns

# Encode the substories and headlines
substory_embeddings = model.encode(data['Sub-Story'].tolist(), convert_to_tensor=True)
headline_embeddings = model.encode(data['HEADLINES'].tolist(), convert_to_tensor=True)

# Function to generate standardized headlines
def generate_standardized_headline(substory):
    substory_embedding = model.encode(substory, convert_to_tensor=True)
    similarities = util.pytorch_cos_sim(substory_embedding, headline_embeddings)
    similarities_np = similarities.cpu().detach().numpy()
    most_similar_idx = np.argmax(similarities_np)
    return data['HEADLINES'].iloc[most_similar_idx]

# Generate headlines for all substories
data['generated_headline'] = data['Sub-Story'].apply(generate_standardized_headline)

# Save the results to an Excel file
output_file = 'substories_and_headlines1.xlsx'
data[['Sub-Story', 'HEADLINES', 'generated_headline']].to_excel(output_file, index=False)

print(f"Excel file '{output_file}' generated successfully.")
```

```python
from nltk.translate.bleu_score import sentence_bleu
import numpy as np

# Calculate BLEU scores for generated headlines
def calculate_bleu_score(reference, candidate):
    # Check for NaN values
    if pd.isna(reference) or pd.isna(candidate):
        return np.nan  # Return NaN if either value is NaN
    
    reference_tokens = reference.split()
    candidate_tokens = candidate.split()
    return sentence_bleu([reference_tokens], candidate_tokens)

# Compute BLEU score for each generated headline
data['bleu_score'] = data.apply(lambda row: calculate_bleu_score(row['HEADLINES'], row['generated_headline']), axis=1)
average_bleu = data['bleu_score'].mean()

# Print the average BLEU score, ensuring to handle NaN values
print(f"Average BLEU Score: {average_bleu:.4f}" if not np.isnan(average_bleu) else "No valid BLEU scores to calculate.")
```

```python
from rouge import Rouge
import numpy as np
import pandas as pd

# Initialize ROUGE
rouge = Rouge()

# Function to filter out NaN values
def filter_nan_values(hyps, refs):
    valid_hyps = []
    valid_refs = []
    for hyp, ref in zip(hyps, refs):
        if pd.notna(hyp) and pd.notna(ref):
            valid_hyps.append(hyp)
            valid_refs.append(ref)
    return valid_hyps, valid_refs

# Filter out NaN values from generated_headline and HEADLINES
filtered_generated, filtered_references = filter_nan_values(data['generated_headline'], data['HEADLINES'])

# Calculate ROUGE scores
rouge_scores = rouge.get_scores(filtered_generated, filtered_references, avg=True)

print("ROUGE Scores:")
print(rouge_scores)
```
